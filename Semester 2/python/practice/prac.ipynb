```python
import numpy as np

np.array([1,2,3])           # create array
np.zeros((2,3))             # zeros
np.ones((2,2))              # ones
np.arange(0,10,2)           # range array
np.linspace(0,1,5)          # evenly spaced

a.shape                     # shape
a.ndim                      # dimensions
a.size                      # number of elements
a.dtype                     # data type

a[0]                        # first element
a[1:4]                      # slice
a[:,1]                      # column
a[0,:]                      # row

a + b                       # add
a * b                       # element-wise multiply
a @ b                       # matrix multiply

a.sum()
a.mean()
a.max()
a.min()

a.reshape(2,3)              # reshape
a.flatten()                 # make 1D

a[a > 5]                    # filtering
```
